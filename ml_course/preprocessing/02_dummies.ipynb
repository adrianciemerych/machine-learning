```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__

data = {
    'size' : ['XL', 'L', 'M', 'L', 'M'],
    'color' : ['red', 'green', 'blue', 'green', 'red'],
    'gender' : ['female', 'male', 'male', 'female', 'female'],
    'price' : [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight' : [500,450,300,380,410],
    'bought' : ['yes', 'no', 'yes', 'no', 'yes'] 
}

df_raw = pd.DataFrame(data)
df_raw

df = df_raw.copy()

for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output = False)
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(sparse_output = False, drop = 'first')
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

```python
df
```

```python
pd.get_dummies(data = df)
```

```python
pd.get_dummies(data = df, drop_first = True)
```

```python
# Dodatkowe ciekawe funkcje
# get_dummies jest bardziej preferowanym łatwiejszym sposobem na 
# zamienienie danych na zera i jedynki
pd.get_dummies(data = df, drop_first = True, columns = ['size'], prefix = 'sizeee', prefix_sep = '___')
```
