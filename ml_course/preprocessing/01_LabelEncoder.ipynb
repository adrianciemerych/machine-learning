Importowanie bibliotek

```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

Wygenerowanie danych

```python
data = {
    'size' : ['XL', 'L', 'M', 'L', 'M'],
    'color' : ['red', 'green', 'blue', 'green', 'red'],
    'gender' : ['female', 'male', 'male', 'female', 'female'],
    'price' : [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight' : [500,450,300,380,410],
    'bought' : ['yes', 'no', 'yes', 'no', 'yes'] 
}

df_raw = pd.DataFrame(data)
df_raw
```

Utworzenie kopii

```python
df = df_raw.copy()
df.info()
```

Zmiana typu danych i wstępna eksploracja

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
# domyślnie wyświetla wyniki tylko dla kolumn numerycznych
```

```python
df.describe().T
df.describe().transpose()
```

```python
# aby wyświetlić statystyki dla innych dtype - parametr 'include'
df.describe(include = 'category').T
```

Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

# klasa LabelEncoder pozwala na zamienienie zmiennej na wartości 1,0
#tworzymy instancję klasy
le = LabelEncoder()
# dopasowanie zmiennej do klasy
le.fit(df['bought'])

# następnie używamy metody transform żeby zmapować dane na takie ktore chcemy
le.transform(df['bought'])

# metoda le.fit_transform(df['bought']) spełnia obie ww. funkcje  
```

```python
# W celu sprawdzenia jak wyglądają klasy
le.classes_
```

```python
le.fit_transform(df['bought'])
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
# Wcelu odwrócenia naniesionych zmian można zastosować 
le.inverse_transform(df['bought'])
```

```python
df['bought'] = le.inverse_transform(df['bought'])
df

# Należy pamiętać! : Klasa LabelEncoder służy do przekształcania naszych 
# zmiennych docelowych. Nie powinno się jej stosować do przekształacania
# zmiennych opisujących
```
